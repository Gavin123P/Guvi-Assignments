```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
```

```python
df1 = pd.read_csv(r"C:\Users\Dell\Documents\Final Project\Data\Sales Forecasting\features.csv")
df2 = pd.read_csv(r"C:\Users\Dell\Documents\Final Project\Data\Sales Forecasting\stores.csv")
df3 = pd.read_csv(r"C:\Users\Dell\Documents\Final Project\Data\Sales Forecasting\train.csv")
df4 = pd.merge(df1, df2, how='outer')
df = pd.merge(df4, df3, how='outer')
#df.head()
df.info()
```

```python
df = df.fillna(df.mean())
df.info()
```

```python
df.head()
```

```python
df['Date'] = pd.to_numeric(pd.to_datetime(df['Date']))
```

```python
df5 = df.replace({'Type':{'A':0,'B':1,'C':2}})
```

```python
df5.head()
```

```python
df5.corr()
```

```python
x = df5.iloc[:,0:-1].values
y = df5.iloc[:,-1:].values
```

```python
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
dt_regressor = DecisionTreeRegressor()
dt_regressor.fit(x_train, y_train)
```

```python
y_pred_dt = dt_regressor.predict(x_test)
```

```python
print(r2_score(y_test, y_pred_dt))
```
