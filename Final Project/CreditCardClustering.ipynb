```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\Dell\Documents\Final Project\Data\Credit Card_Clustering\Credit Card_Clustering.csv")
X = data.iloc[:, 1:]
X = X.fillna(X.mean())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=5, random_state=0, n_init=10).fit(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis')
plt.show()
```
